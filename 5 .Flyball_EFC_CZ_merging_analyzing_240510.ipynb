```python
# 20240329 ez jó szöveges dátumot csinál a 'when_date'-be
import pandas as pd
import re

def extract_dates(df):
    # Ellenőrizzük, hogy van-e "event_place" oszlop
    if 'event_place' not in df.columns:
        print("Error: 'event_place' column not found in DataFrame.")
        return None

    # Az "event_place" oszlop feldolgozása
    df['when_date'] = df['event_place'].apply(lambda x: x[x.find('2023'):].strip() if '2023' in str(x) else None)

    return df

# Tesztelés
#df_EFC_CZ_merged = pd.read_csv("df_EFC_CZ_merged.csv")
df_with_dates = extract_dates(df_EFC_CZ_cleaned)
di(df_with_dates.head(30))
```

```python
#240329 ez jó számos dátumot csinál
import pandas as pd

def convert_date_format(date_string):
    # A magyar hónapok angol megfelelői
    month_translations = {
        'január': '01',
        'február': '02',
        'március': '03',
        'április': '04',
        'május': '05',
        'június': '06',
        'július': '07',
        'augusztus': '08',
        'szeptember': '09',
        'október': '10',
        'november': '11',
        'december': '12'
    }

    # A dátum string feldarabolása a kötőjel mentén
    parts = date_string.split('-')

    # Az első rész tartalmazza az eredeti dátumot
    original_date = parts[0].strip()

    # Az első rész további részeinek feldarabolása
    original_date_parts = original_date.split()
    if len(original_date_parts) == 3:
        # Az év, hónap és nap kinyerése
        year = original_date_parts[0].strip('.')
        month = original_date_parts[1].strip().lower()  # Hónap kisbetűsítése
        day = original_date_parts[2].strip('.')

        # Hónapok fordítása szám formátumra
        if month in month_translations:
            month = month_translations[month]

        # Dátum formátum szerinti értelmezése
        try:
            parsed_date = pd.to_datetime(f"{year}-{month}-{day}")
            return parsed_date.strftime('%Y-%m-%d')
        except ValueError:
            print(f"Failed to parse date: {original_date}")
            return None
    else:
        print(f"Failed to parse date: {original_date}")
        return None

# 'when_date' átalakítása
df_with_dates['when_date'] = df_with_dates['when_date'].apply(convert_date_format)
di(df_with_dates.head(20))
di(df_with_dates.tail(20))
```

```python
# 240330 ez a jó kék-piros pálya felcimkézés, logikailag ugyan nem ide tartozik, de már nincs időm más helyre áthelyezni
def add_track_column(df, rows):
    # Ellenőrizzük, hogy van-e "when_time" oszlop
    if 'when' not in df.columns:
        print("Error: 'when' column not found in DataFrame.")
        return None

    # "track" oszlop inicializálása üres listával
    track_column = []

    # Az egyedi dátum-idő párok kigyűjtése az adott sorokra korlátozva
    unique_datetime_pairs = df.loc[rows, 'when'].unique()

    # Az egyedi dátum-idő párokhoz tartozó "track" értékek meghatározása
    for datetime_pair in unique_datetime_pairs:
        # Az aktuális dátum-idő párhoz tartozó indexek lekérdezése
        indices = df.index[(df.loc[rows, 'when'] == datetime_pair)].tolist()

        # Az azonos dátum-idő párokhoz tartozó "track" érték meghatározása
        track = None
        for i, index in enumerate(indices):
            if track is None:
                track = "kék"
            else:
                track = "piros"
            track_column.append(track)

    # Ellenőrizzük a rows változó hosszát
    print(len(rows))
    # Ellenőrizzük, hogy a track_column hossza egyezik-e a DataFrame hosszával
    if len(track_column) != len(rows):
        print("Error: Length of 'track_column' does not match length of specified rows.")
        return None

    # "track" oszlop hozzáadása a DataFrame-hez az adott sorokon belül
    df.loc[rows, 'track'] = track_column

    return df

# Az adott sorok kiválasztása
rows = df_with_dates['when'].notna()

# "track" oszlop hozzáadása a DataFrame-hez az adott sorokon belül
df_EFC_CZ_track = add_track_column(df_with_dates, rows)

# Eredmény kiíratása és megnézem mindenhova csak ez a két érték került
li(df_EFC_CZ_track,20)
luv(df_EFC_CZ_track, columns=['track'])
```

####Látszik, hogy "when_date" a sok helyen kötőjeles, két napos dátumoknál csak az első dátumot tartalmaza. Ezt az alábbi függvény hivatott javítani.


Ez a kód egy DataFrame-en (df) végigmegy, és ellenőrzi az időpontokat. Ha az aktuális időpont (current_time) kisebb, mint az előző időpont (previous_time) ugyanazon a napon (previous_date==current_date), akkor kiírja az aktuális és az előző időpontot, majd hozzáad egy napot az aktuális dátumhoz. Ezután frissíti az when_date oszlopot a DataFrame-ben az új dátummal (new_date).

Ha az aktuális időpont nagyobb, mint az előző időpont, akkor az aktuális sor dátumát az előző sor dátumára állítja vissza.

```python
#240419 most jó lett, a "when_date"-t korrigálja más oszlopok változásinak függvényében.  (végig ellenőriztem hibátlan)
from datetime import datetime, timedelta

def check_and_update_dates(df):
    previous_time = None
    previous_date = None
    previous_place = None
    for index, row in df.iterrows():
        current_time = row['when_time']
        current_date = row['when_date']
        current_place = df.at[index, 'event_place']
        if previous_time is not None and current_time < previous_time and previous_date==current_date and index > 1061:
            print(f"A(z) curr:{current_date}  prev:{previous_date}    dátumhoz tartozó idő kisebb, mint az előző időpont   cur:{current_time}  prev:{previous_time}   {current_place}.")
            datum = datetime.strptime(current_date, "%Y-%m-%d")
            uj_datum = datum + timedelta(days=1)
            current_date = uj_datum.strftime("%Y-%m-%d")
            print(current_date)

            new_date = current_date
            df.at[index, 'when_date'] = new_date
        elif previous_time is not None and current_time >= previous_time and previous_place==current_place:

            df.at[index, 'when_date'] = df.at[index - 1, 'when_date']
        previous_time = current_time
        previous_date = current_date
        previous_place = current_place

check_and_update_dates(df_EFC_CZ_track)

#di(df_EFC_CZ_track.tail(10))
```

```python
import numpy as np  # NumPy importálása
from scipy import stats  # Scipy importálása a statisztikai műveletekhez

def numeric_column_summary_statistics_full(df):
    # Az összes oszlop fejlécének kiválasztása
    all_columns = df.columns

    # Csak a számot tartalmazó oszlopok kiválasztása
    numeric_columns = df.select_dtypes(include='number').columns

    # Statisztikák kiszámolása minden oszlopra
    statistics = pd.DataFrame(index=['Min', 'Max', 'Átlag', 'Medián', 'Modusz', 'Variabilitás', 'Alsó Kvantil', 'Felső Kvantil',
                                     'Korreláció', 'Kovariancia', 'Százalékos Változás', 'Konfidencia-intervallumok',
                                     'Lineáris Regresszió Együtthatók'], columns=all_columns)

    for column in all_columns:
        if column in numeric_columns:
            statistics.at['Min', column] = df[column].min()
            statistics.at['Max', column] = df[column].max()
            statistics.at['Átlag', column] = df[column].mean()
            statistics.at['Medián', column] = df[column].median()
            statistics.at['Modusz', column] = df[column].mode().iloc[0]  # Modusz számítása
            statistics.at['Variabilitás', column] = df[column].std()  # Szórás
            statistics.at['Alsó Kvantil', column] = df[column].quantile(0.25)  # Alsó kvantil
            statistics.at['Felső Kvantil', column] = df[column].quantile(0.75)  # Felső kvantil
            # Korreláció és kovariancia kiszámítása a többi oszloppal
            for other_column in numeric_columns:
                if other_column != column:
                    statistics.at['Korreláció', column] = df[column].corr(df[other_column])
                    statistics.at['Kovariancia', column] = df[column].cov(df[other_column])
            # Százalékos változás számítása
            first_value = df[column].iloc[0]
            last_value = df[column].iloc[-1]
            statistics.at['Százalékos Változás', column] = ((last_value - first_value) / first_value) * 100
            # Konfidencia-intervallumok
            confidence_interval = stats.norm.interval(0.95, loc=df[column].mean(), scale=df[column].std())
            statistics.at['Konfidencia-intervallumok', column] = confidence_interval
            # Lineáris regresszió együtthatók
            # Ez csak egy példa, itt valószínűleg több munka és elemzés kellene
            slope, intercept, r_value, p_value, std_err = stats.linregress(df[column], df['start'])
            statistics.at['Lineáris Regresszió Együtthatók', column] = (slope, intercept, r_value, p_value, std_err)
        else:
            # Nem szám típusú oszlopoknak NaN értékeket adunk
            statistics.at['Min', column] = np.nan
            statistics.at['Max', column] = np.nan
            statistics.at['Átlag', column] = np.nan
            statistics.at['Medián', column] = np.nan
            statistics.at['Modusz', column] = np.nan
            statistics.at['Variabilitás', column] = np.nan
            statistics.at['Alsó Kvantil', column] = np.nan
            statistics.at['Felső Kvantil', column] = np.nan
            statistics.at['Korreláció', column] = np.nan
            statistics.at['Kovariancia', column] = np.nan
            statistics.at['Százalékos Változás', column] = np.nan
            statistics.at['Konfidencia-intervallumok', column] = np.nan
            statistics.at['Lineáris Regresszió Együtthatók', column] = np.nan

    # Statisztikák kiíratása
    print("Összefoglaló statisztikák minden oszlopra:")
    display(statistics)

numeric_column_summary_statistics_full(df_EFC_CZ_track)
```
