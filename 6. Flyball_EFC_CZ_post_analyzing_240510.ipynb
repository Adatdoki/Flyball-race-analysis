```python
# helyreteszünk néhány adattípust
def convert_data_types(df, columns, dtype):
    """
    Az adatkeret oszlopainak adattípusát átalakítja a megadott típusra.

    Args:
        df (DataFrame): Az adatkeret, amelynek az oszlopait átalakítani szeretnénk.
        columns (list): Azoknak az oszlopoknak a listája, amelyeket átalakítani szeretnénk.
        dtype (str): Az átalakítandó adattípus.

    Returns:
        DataFrame: Az átalakított adatkeret.
    """
    for col in columns:
        if dtype == "datetime":
            df[col] = pd.to_datetime(df[col], errors='coerce')
        else:
            df[col] = df[col].astype(dtype)

    return df


convert_data_types(df_EFC_CZ_merged, ["1_dog", "chng","2_dog","chng_1","3_dog","chng_2","chng_2"], "float64")
convert_data_types(df_EFC_CZ_merged, ["when_date", "when_time"], "datetime")
```

```python
# 240331 a szöveges adatokból számos oszlopok készülnek
#Ez a függvény létrehoz oszlopokat a DataFrame-ben az adott kategórikus oszlop egyedi értékeinek megfelelően, ahol az új oszlopok binárisak,
#és jelzik, hogy az eredeti oszlop egyes sorai megegyeznek-e az adott egyedi értékkel vagy sem.
def create_binary_columns(dataframe, categorical_columns):
    """
    Létrehoz oszlopokat a DataFrame-ben az adott kategórikus oszlop egyedi értékeinek megfelelően,
    ahol az új oszlopok binárisak, és jelzik, hogy az eredeti oszlop egyes sorai megegyeznek-e az adott egyedi értékkel vagy sem.

    Args:
        dataframe (DataFrame): Az adatkeret, amelyben az oszlopokat létre akarjuk hozni.
        categorical_columns (list): Azon oszlopok listája, amelyek alapján az új bináris oszlopokat létre akarjuk hozni.

    Returns:
        DataFrame: Az új oszlopokat tartalmazó adatkeret.
    """
    new_dataframe = dataframe.copy()  # Készítünk egy másolatot az eredeti adatkeretről

    for column in categorical_columns:
        for value in dataframe[column].unique():
            def f(x):
                if x == value:
                    return 1
                else:
                    return 0
            new_dataframe[column + "_" + str(value)] = new_dataframe[column].apply(f)

    return new_dataframe


df_EFC_CZ_merged = create_binary_columns(df_EFC_CZ_merged, ('W/L/T', 'track'))
li(df_EFC_CZ_merged.head(20))
```

```python
def get_dog_names(df, column_names):
    dog_names_set = set()  # Kutyanevek halmaza
    for column_name in column_names:
        dog_names_set.update(df[column_name].dropna().unique())

    dog_names_list = sorted(list(dog_names_set))  # Kutyanevek listába rendezve

    # Kutyanevekhez tartozó sorok kinyerése
    dog_rows = []
    for dog_name in dog_names_list:
        dog_rows.extend(df[df[column_names].eq(dog_name).any(axis=1)].index)

    num_unique_dog_names = len(dog_names_list)
    return dog_names_list, df.loc[dog_rows], num_unique_dog_names

# Tesztelés
dog_names_list, dog_rows_df, num_unique_dog_names = get_dog_names(df_EFC_CZ_merged, ["name", "name_1", "name_2", "name_3"])
print("Kutya nevek listája:")
print(dog_names_list)
print("\nÖsszesen {} kutyanevet találtam.".format(num_unique_dog_names))
```

```python
import pandas as pd

# Kutyanevek kinyerése az összes oszlopból
kutyanevek = df_EFC_CZ_merged[['name', 'name_1', 'name_2', 'name_3']].values.ravel()
kutyanevek = [nev for nev in kutyanevek if pd.notna(nev)]
egyedi_kutyanevek = sorted(set(kutyanevek))

# Nyert kutyák kék és piros pályán statisztikája
nyert_kutya_kék_pályán = df_EFC_CZ_merged[df_EFC_CZ_merged['nyert_kék_pályán'] == True]
nyert_kutya_piros_pályán = df_EFC_CZ_merged[df_EFC_CZ_merged['nyert_piros_pályán'] == True]

nyert_kutya_kék_pályán_statisztika = nyert_kutya_kék_pályán.groupby('name').size()
nyert_kutya_piros_pályán_statisztika = nyert_kutya_piros_pályán.groupby('name').size()

# Kutyanevek DataFrame létrehozása
kutyanevek_df = pd.DataFrame({
    'Kék pálya': nyert_kutya_kék_pályán_statisztika,
    'Piros pálya': nyert_kutya_piros_pályán_statisztika
})

# Üres értékek helyettesítése nullával
kutyanevek_df.fillna(0, inplace=True)

# Index (kutyanev) beállítása
kutyanevek_df.index.name = 'Kutya neve'

# Csapatok neveinek hozzáadása a kutyanevek DataFrame-hez
for index, row in df_EFC_CZ_merged.iterrows():
    csapat = row['who']
    for col in ['name', 'name_1', 'name_2', 'name_3']:
        kutya_nev = row[col]
        if pd.notna(kutya_nev) and pd.notna(csapat):
            kutyanevek_df.loc[kutya_nev, 'Csapat'] = csapat

# Eredmény megjelenítése
print(kutyanevek_df)
```

```python
from scipy.stats import ttest_ind

# Szignifikanciaszint
alpha = 0.05

# Adatok előkészítése
kék_pálya = [42.0, 30.0, 20.0, 12.0, 10.0, 8.0, 7.0, 6.0, 6.0, 6.0, 5.0, 5.0, 1.0, 1.0, 1.0]
piros_pálya = [36.0, 24.0, 23.0, 18.0, 12.0, 11.0, 10.0, 8.0, 8.0, 6.0, 6.0, 6.0, 3.0, 3.0, 2.0, 2.0, 1.0, 0.0]

# T-próba teszt végrehajtása
t_stat, p_value = ttest_ind(kék_pálya, piros_pálya, nan_policy='omit')

# Eredmények megjelenítése
print("T-próba statisztika:", t_stat)
print("p-value:", p_value)

# P-value érték értékelése
if p_value < alpha:
    print("Van szignifikáns különbség a két csoport között.")
else:
    print("Nincs szignifikáns különbség a két csoport között.")
```

```python
def get_dog_names(df, column_names):
    dog_names_set = set()  # Kutyanevek halmaza
    for column_name in column_names:
        dog_names_set.update(df[column_name].dropna().unique())

    dog_names_list = sorted(list(dog_names_set))  # Kutyanevek listába rendezve

    # Kutyanevekhez tartozó sorok kinyerése
    dog_rows = []
    for dog_name in dog_names_list:
        dog_rows.extend(df[df[column_names].eq(dog_name).any(axis=1)].index)

    num_unique_dog_names = len(dog_names_list)
    return dog_names_list, df.loc[dog_rows], num_unique_dog_names

# Tesztelés
dog_names_list, dog_rows_df, num_unique_dog_names = get_dog_names(df_EFC_CZ_merged, ["name", "name_1", "name_2", "name_3"])
print("Kutya nevek listája:")
print(dog_names_list)
print("\nÖsszesen {} kutyanevet találtam.".format(num_unique_dog_names))
```
